#### Application Settings

```python
enable_trueRandom = False
num_games = 25
per_game  = 50_000
```

#### Static Imports

```python
from multimulti_set import MultiMultiSet as Multi
from utils.random_provider import pseudo_sample
from utils.random_provider import true_sample
from utils.export_csv import export_csv
```

## Ticket Generation
Generate tickets using the selected sampling method.

```python
tickets = []
sampler = true_sample if enable_trueRandom else pseudo_sample

for i in range(num_games):
    draw = sampler(range(1, Multi.POOL_SIZE + 1), Multi.DRAW_COUNT)

    for j in range(per_game):
        ticket = Multi(
            i, j,
            sampler(range(1, Multi.POOL_SIZE + 1), Multi.BET_COUNT),
            draw
        )
        tickets.append(ticket)
```

Check for hits, calculate the prize

```python
for t in tickets:
    hits = len(set(t.numbers) & set(t.drawn))
    t.hits = hits
    t.prize = Multi.PRIZE_TABLE[hits]
```

Save to file

```python
rows = [t.serialise() for t in tickets]
export_csv(rows, "data/rolls.csv")
```
